```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
filename = 'reg_unificados_intc.mat.log'
```

```python
df = pd.read_csv(filename, sep=';')
```

```python
df.shape
```

```python
df.sample(5)
```

```python
df['peak_scale'] =  np.where(df['peak'] <0.25, df['peak'] / (93/0.076382), df['peak'] / (373/0.255276))

y=[]
for i in range(359):
    y.append(np.sin(2/360*i*np.pi)*df['peak_scale'].max())

```

```python
fig, ax = plt.subplots(figsize=(8, 6))

plt.title(filename)
plt.xlabel('degree')
plt.ylabel('adc value')
plt.grid()

ax.scatter(df['degree'], df['peak_scale'], color='red', s=1)
ax.plot(range(359), y, color='blue')
plt.savefig(filename + '.png')
plt.show()
```
